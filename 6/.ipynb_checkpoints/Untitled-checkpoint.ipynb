```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
N=10**6
num_mc_iters = 10**3
probabilities = [10**-4, 5*10**-4, 10**-3, 5*10**-3, 10**-2, 5*10**-2, 10**-1]
# probabilities = [10**-2, 5*10**-2, 10**-1]
```

```python
# def make_batches(N, k):
#     if N%k == 0:
#         return np.arange(1, N+1).reshape(N//k, k)
#     else:
#         batches = np.arange(1, N+1)
#         extra_to_add = [0]*(k - N%k)
#         batches = np.append(batches, extra_to_add)
        
#         return batches.reshape(N//k, k)
    

def find_multiples(num):
    multiples = []
    for i in range(1, num+1):
        if num%i == 0:
            multiples.append(i)
            
    return multiples
```

```python
## Problem with making batches approach: we are padding with zeros to make a matrix but this will change the
## size of the batch if N%k != 0.
```

```python
p_batch_and_num_tests = dict()

for p in probabilities:
    batch_and_num_tests = dict()

#     for k in range(0, N+1, 10):
#         if k == 0:
#             k = 1
    
    for k in find_multiples(N):
        num_tests = 0
        for _ in range(num_mc_iters):
            infected_set = set(random.sample(range(1, N+1), int(p*N)))
            
            def check_infected(batch):
                return len(batch) if len(set(batch).intersection(infected_set)) > 0 else 0
            
            
            batches = np.arange(1, N+1).reshape(N//k, k)
            result = sum(np.array(list(map(check_infected, batches))))
            num_tests += result
            num_tests += batches.shape[0]

#             for i in range(1, N+1, k):
#                 end = min(N, i+k-1)
#                 batch = set(np.arange(i, end+1))
                
#                 is_batch_infected = True if len(batch.intersection(infected_set)) > 0 else False

#                 if is_batch_infected:
#                     num_tests += len(batch)

#                 num_tests += 1

        batch_and_num_tests[k] = num_tests/num_mc_iters
    
    p_batch_and_num_tests[p] = batch_and_num_tests
```

```python
plt.figure(figsize=(15, 15))

for p, p_data in p_batch_and_num_tests.items():
    plt.plot(p_data.keys(), p_data.values(), label=p)
    
plt.xlim(0, 100)
plt.xlabel("Batch size")
plt.ylabel("Num tests conducted")
plt.legend()
```
